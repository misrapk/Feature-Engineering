# Technique 3 --> For new feature

It works well if the data are not missing completely at random.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("E:\\100DaysDataScience\\datasets\\titanic_train.csv", usecols=['Age', 'Fare', 'Survived'])
```

```python
df.head()
```

```python
# Create a new feature and wherever there is null than replace with 
# 1 else replace with 0

df['Age_NAN'] = np.where(df['Age'].isnull() , 1,0)
```

```python
df.head()
```

```python
df.head(20)
```

With this we provide addtional feature for NaN value
and now we can replace the values of 1 with either mean/median/mode

### Advantages/Disadvantages of Capturing NAN

##### 1. Advantages

``--> Easy to Implement(Robust to Outliers)``

``--> captures the importance of missing values``

##### 2. DisAdvantages

``--> THis method is creating additional feature`` This may lead to Curse of Dimensionality
