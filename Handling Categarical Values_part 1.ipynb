### IF you have many categories than How can i Perform One hot

```python
data = pd.read_csv('mercedese.csv')
```

```python
data.head()
```

```python
#lets take first 6 columns
df = pd.read_csv('mercedese.csv', usecols=['X0', 'X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
for i in df.columns:
    print(len(df[i].unique()))
```

So here we can see there are so many categories there.

So in this ....if we use **one hot encoding** then It'll be a big challange.

**For this .... we will use the method that was used in KDD orannge cup challange**
##### In this we will take 10 most frequent values and perfoorm one hot encoding on it

```python
# take top 10 most frequent occuring variables

df.X1.value_counts().sort_values(ascending=False).head(10)
```

```python
# lets perform the One Hot Encoding in it
top10VariableIndex = df.X1.value_counts().sort_values(ascending=False).head(10).index
top10VariableIndex = list(top10VariableIndex)
top10VariableIndex
```

```python
# now perform on this
for categories in top10VariableIndex:
    df[categories] = np.where(df['X1'] == categories, 1, 0)  
    #where the categories is present it will replace with 1 else 0
```

```python
df[top10VariableIndex].head(10)
```
